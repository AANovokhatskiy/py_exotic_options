```python
import numpy as np
from matplotlib import pyplot as plt
from european_call import european_call_option
from barrier_call import barrier_call_option
from lookback_call import lookback_call_option
```

European call

```python
'''Parameters'''
n_t, n_s = 4000, 5000
n_mc = int(10**5)
T, S0, K, r, sigma = 1, 15, 16, 0.07, 0.05

option = european_call_option(T, 0, S0, K, r, sigma)

mc_price = option.price_monte_carlo(n_mc)
exact_price = option.price_exact()
t, s, v = option.price_pde(n_t, n_s)
pde_price = option.get_pde_result()

'''Plot'''
s, t = np.meshgrid(s, t)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(s, t, v, cmap="plasma")
ax.set_xlabel('S')
ax.set_ylabel('t')
ax.set_zlabel('V(S,t)')
ax.view_init(30, 250)
plt.show()

'''Result'''
print("European call")
print(f"Parameters T = {T}, S0 = {S0}, K = {K}, r = {r}, sigma = {sigma}")
print(f"Grid n x m = {n_t:.1E} x {n_s:.1E}")
print(f"Number of Monte Carlo iterations = {n_mc:.1E}")
print(f"Exact price = {exact_price:.9f}")
print(f"MC price = {mc_price:.9f}")
print(f"PDE price = {pde_price:.9f}")
```
